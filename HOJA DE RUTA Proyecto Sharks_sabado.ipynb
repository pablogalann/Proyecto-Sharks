Creamos y aplicamos una función que repita un patrón de limpieza y nos devuelva datos limpios, consistentes y enteros

```python
def limpiar_Age(x):
    
    if ',' in x:
        x = x.split(' , ')
        return  x[0]
    

    if '&' in x:
        x = x.split(' & ')
        return x[1]
    
    if 's' in x:
        return x[1]
    
    if 'teen' in x.lower():
        return 15
    
    if 'or' in x:
        x = x.split(' or ')
        return x[1]
    
    if 'adult' in x:
          return 35
          
    if 'Elderly' in x:
          return 70

    if 'young' in x:
          return 22
        
    if 'middle-age' in x:
          return 22
        
    if 'months' in x:
          return 1
    
    if ' ' in x:
          return 'Unknown'
        
    if '?' in x:
          return 'Unknown'
        
    return x
        
```

```python
def limpiar_Time(x):
    
    if '/' in x:
        x = x.split(' / ')
        return  x[0]
    
    if 'or' in x:
        x = x.split(' or ')
        return  x[0]
    
    if '--' in x:
        x = x.split(' -- ')
        return  x[0]
    
    if 'before' in x:
        x = x.split(' before ')
        return  x[1]
    
    if '-' in x:
        x = x.split(' - ')
        return  x[0]
    
    if 'afternoon' in x.lower():
        return  '19h00'
    
    if 'morning' in x.lower():
        return  '10h00'
    
    if 'night' in x.lower():
        return  '23h00'
    
    if 'night' in x.lower():
        return  '23h00'
    
    if 'evening' in x.lower():
        return  '21h00'
    
    if 'dusk' in x.lower():
        return  '20h00'
    
    if 'midday' in x.lower():
        return  '12h00'
    
    if 'early_afternoon' in x.lower():
        return  '16h00'
    
    if 'sunset' in x.lower():
        return  '20h00'
    
    if 'midnight' in x.lower():
        return  '00h00'
    
    if 'dawn' in x.lower():
        return  '07h00'
    
    if 'dunk' in x.lower():
        return  '20h00'
   
        
    return x
```
